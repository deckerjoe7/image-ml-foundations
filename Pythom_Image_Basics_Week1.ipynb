```python
# Load and display
import cv2
from PIL import Image
import matplotlib.pyplot as plt
import numpy as np

uploaded = '/content/drive/MyDrive/2025/Image ML Foundations/IMG_4694.jpeg'

# setting up figure
plt.figure(figsize=(12, 6))


# With openCV
img_cv = cv2.imread(uploaded)
# Check if the image was loaded successfully
if img_cv is None:
    print(f"Error: Could not read image from {uploaded}. Please check the file path.")
else:
    img_rgb = cv2.cvtColor(img_cv, cv2.COLOR_BGR2RGB) # convert from BGR to RGB
    plt.subplot(1,2,1)
    plt.imshow(img_rgb)
    plt.title('Using CV')


# With PIL
img_pil = Image.open(uploaded)
plt.subplot(1,2,2)
plt.imshow(img_pil)
plt.title('Using PIL')
```

```python
# Common image transformations
img_resized = cv2.resize(img_rgb, (224,224))
img_gray = cv2.cvtColor(img_resized, cv2.COLOR_RGB2GRAY) # convert from BGR to RGB

plt.figure(figsize=(12,8))
plt.subplot(1,3,1)
plt.imshow(img_rgb)
plt.subplot(1,3,2)
plt.imshow(img_resized)
plt.subplot(1,3,3)
plt.imshow(img_gray, cmap='gray')
```

```python
# Cropping
h,w,i = img_rgb.shape
img_crop = img_rgb[h//4:3*h//4,w//4:3*w//4]

plt.imshow(img_crop)
```
